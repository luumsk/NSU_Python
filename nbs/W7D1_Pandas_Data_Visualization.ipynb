# Data Visualization with Pandas

- The main goal of data visualization is to make the readers identify patterns, trends and outliers easily in large data sets.

- Data visualization is a crucial component of data analysis, and Pandas provides some basic tools for visualizing data directly from DataFrames.

- Pandas supports various plot types, including line plots, bar plots, scatter plots, histograms, box plots, and more. The choice of plot type depends on the data and what you want to visualize.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("../data/billionairs.csv")
```

```python
df.describe()
```

```python
countries = ["Finland", "Norway"]
df[df["country"].isin(countries)]
```

## 1. Histogram

## 2. Line plot

## 3. Scatter plot

```python
df = pd.read_csv("../data/billionairs.csv")
```

```python
df.head()
```

```python
df.plot(kind="scatter", x="age", y="finalWorth", figsize=(15, 5))
```

## 4. Bar plot

```python
gb = df.groupby("country").agg({"finalWorth": "sum"}).sort_values("finalWorth")[:10]
gb
```

```python
gb.plot(kind="barh")
```

## 5. Pie plot

```python
ax = df.value_counts("gender").plot(kind="pie")
```
